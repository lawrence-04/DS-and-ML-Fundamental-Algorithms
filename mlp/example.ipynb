```python
import numpy as np
import matplotlib.pyplot as plt

from mlp import MLP
```

```python
# generate a simple synthetic regression dataset
np.random.seed(42)
num_samples = 200
num_features = 2
X = np.random.randn(num_features, num_samples)

# True weights and bias for a linear function
true_W = np.array([[2.0, -3.5]])
true_b = np.array([[0.5]])

# Generate targets with some noise
y = true_W @ X + true_b + 0.1 * np.random.randn(1, num_samples)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].scatter(X[0, :], y[0])
axes[1].scatter(X[1, :], y[0])

axes[0].set_ylabel("target")
axes[0].set_xlabel("feature 1")
axes[1].set_ylabel("target")
axes[1].set_xlabel("feature 2")
```

```python
# Initialize and train MLP
mlp = MLP(num_inputs=num_features, num_hidden=5, num_outputs=1, lr=0.1)
mlp.fit(X, y, epochs=50)
```

```python
plt.plot(mlp.losses)
plt.ylabel("loss")
plt.xlabel("epoch")
```
