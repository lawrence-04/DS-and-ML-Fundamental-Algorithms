```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from tf_idf import vectorise_corpus
```

```python
reviews = [
    "This laptop is amazing! Super fast and reliable.",
    "Worst purchase ever. The screen broke in a week.",
    "It's okay, not the best but not terrible either.",
    "Love the battery life and the design. Very sleek!",
    "Support was unhelpful and rude. Not buying again.",
]
```

```python
tf_idf_vectors = vectorise_corpus(reviews)
tf_idf_dataframe = pd.DataFrame(tf_idf_vectors).fillna(0)
tf_idf_dataframe.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(tf_idf_dataframe, ax=ax)
```
