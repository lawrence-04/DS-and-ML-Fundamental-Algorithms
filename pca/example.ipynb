```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pca import PCA
```

```python
# Generate a football-shaped cloud
np.random.seed(42)
mean = [0, 0]
cov = [[3, 2], [2, 2]]  # Elongated, tilted shape due to correlation
X = np.random.multivariate_normal(mean, cov, 300)

fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(x=X[:, 0], y=X[:, 1], ax=ax)
ax.set_title("Raw Data");
```

```python
# Apply PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(x=X_reduced[:, 0], y=X_reduced[:, 1], ax=ax)
ax.set_title("Principle Components");
```
